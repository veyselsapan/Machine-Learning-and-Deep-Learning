```python
import re

# Example text
text = "The rain in Spain stays mainly in the plain."

# Search: Find the first occurrence of the word 'rain'
search_result = re.search(r"rain", text)
if search_result:
    print(f"Found '{search_result.group()}' at index {search_result.start()}")

# Match: Check if the text starts with "The"
match_result = re.match(r"The", text)
if match_result:
    print(f"Text starts with '{match_result.group()}'")

# Findall: Find all words ending with 'ain'
all_results = re.findall(r"\b\w+ain\b", text)
print("Words ending with 'ain':", all_results)

# Sub: Replace 'rain' with 'shine'
replaced_text = re.sub(r"rain", "shine", text)
print("Replaced text:", replaced_text)

# Split: Split the text on every space
split_text = re.split(r"\s+", text)
print("Split text:", split_text)
```

```python
import re

# Example 1: Matching one of two words
pattern = r"apple|banana"
text = "I like apple and banana."
matches = re.findall(pattern, text)
print("Example 1 - Matches:", matches)

# Example 2: Matching with grouping
pattern = r"(apple|banana)s"
text = "I see apples, bananas, and grapes."
matches = re.findall(pattern, text)
print("Example 2 - Matches:", matches)

# Example 3: Complex pattern with grouping and disjunction
pattern = r"(colou?r|gray)"
text = "The color is gray, but sometimes colour can also be gray."
matches = re.findall(pattern, text)
print("Example 3 - Matches:", matches)

# Example 4: Using disjunction for patterns with varying prefixes
pattern = r"(pre|sub)fix"
text = "prefix, postfix, and subfix are related."
matches = re.findall(pattern, text)
print("Example 4 - Matches:", matches)

# Example 5: Optional prefix with disjunction
pattern = r"(\d+ (miles|kilometers))"
text = "We drove 100 miles, then 50 kilometers, and finally 80 miles."
matches = re.findall(pattern, text)
print("Example 5 - Matches:", matches)
```

```python
import re

# Example 1: Negated Character Class
pattern = r"[^aeiou]"
text = "Regular expressions can match patterns."
matches = re.findall(pattern, text)
print("Example 1 - Non-vowel characters:", matches)

# Example 2: Negative Lookahead
pattern = r"a(?!b)"
text = "abacada"
matches = re.findall(pattern, text)
print("Example 2 - 'a' not followed by 'b':", matches)

# Example 3: Negative Lookbehind
pattern = r"(?<!b)a"
text = "abacada"
matches = re.findall(pattern, text)
print("Example 3 - 'a' not preceded by 'b':", matches)

# Example 4: Combining Negative Lookahead with Other Patterns
pattern = r"\b(?!cat)\w+"
text = "I like cats and dogs."
matches = re.findall(pattern, text)
print("Example 4 - Words not starting with 'cat':", matches)
```

```python
import re

# Example 1: Single Optional Character
pattern = r"colou?r"
text = "Do you write it as color or colour?"
matches = re.findall(pattern, text)
print("Example 1 - Matches:", matches)

# Example 2: Optional Group
pattern = r"a(bc)?d"
text = "ad, abcd, and aefg"
matches = re.findall(pattern, text)
print("Example 2 - Matches:", matches)

# Example 3: Non-Greedy Quantifier
pattern = r"<.*?>"
text = "<html><body>Hello</body></html>"
matches = re.findall(pattern, text)
print("Example 3 - Non-greedy Matches:", matches)

# Example 4: Multiple Optional Characters
pattern = r"https?://(www\.)?\w+\.\w+"
text = "Visit our site at http://example.com or https://www.example.org."
matches = re.findall(pattern, text)
print("Example 4 - Matches:", matches)
```

```python
import re

# Example 1: Named Group with Aliases
pattern = r"(?P<protocol>https?)://(?P<subdomain>www\.)?(?P<domain>\w+\.\w+)"
text = "Check out http://www.example.com or https://example.org."
matches = re.finditer(pattern, text)
for match in matches:
    print("Example 1 - Protocol:", match.group("protocol"))
    print("Subdomain:", match.group("subdomain") or "N/A")
    print("Domain:", match.group("domain"))

# Example 2: Backreference Using Named Groups
pattern = r"(?P<word>\b\w+\b) (?P=word)"
text = "hello hello world"
matches = re.findall(pattern, text)
print("Example 2 - Repeated Words:", matches)

# Example 3: Breaking Down Complex Expressions
# Define smaller reusable patterns
protocol = r"https?"
subdomain = r"(www\.)?"
domain = r"\w+\.\w+"

# Combine them into a single regex pattern
pattern = fr"{protocol}://{subdomain}{domain}"
text = "Visit our site at http://example.com or https://www.example.org."
matches = re.findall(pattern, text)
print("Example 3 - Combined Pattern Matches:", matches)
```

```python
import re

# Example 1: ^ (Start of a Line/String)
pattern = r"^hello"
text = "hello world!\nhello again!"
matches = re.findall(pattern, text, re.MULTILINE)
print("Example 1 - Matches (Start of Line):", matches)

# Example 2: $ (End of a Line/String)
pattern = r"again!$"
text = "hello world!\nhello again!"
matches = re.findall(pattern, text, re.MULTILINE)
print("Example 2 - Matches (End of Line):", matches)

# Example 3: \b (Word Boundary)
pattern = r"\bworld\b"
text = "hello world! hello world-class!"
matches = re.findall(pattern, text)
print("Example 3 - Matches (Word Boundary):", matches)

# Example 4: \B (Non-Word Boundary)
pattern = r"\Bworld\B"
text = "hello swordfish!"
matches = re.findall(pattern, text)
print("Example 4 - Matches (Non-Word Boundary):", matches)

# Example 5: \A (Start of the Entire String)
pattern = r"\Ahello"
text = "hello world!\nhello again!"
match = re.search(pattern, text)
print("Example 5 - Match (Start of Entire String):", match.group() if match else "No match")

# Example 6: \Z (End of the Entire String)
pattern = r"again!\Z"
text = "hello world!\nhello again!"
match = re.search(pattern, text)
print("Example 6 - Match (End of Entire String):", match.group() if match else "No match")
```

```python
import nltk
from nltk.corpus import brown

# Download required NLTK datasets
nltk.download('brown')

# Step 1: Create a list of (condition, event) pairs.
# In this case, the condition is the genre, and the event is the word.
pairs = [(genre, word.lower()) for genre in brown.categories() for word in brown.words(categories=genre)]

# Step 2: Construct the Conditional Frequency Distribution
cfd = nltk.ConditionalFreqDist(pairs)

# Step 3: Inspect the results
# Example 1: Frequency distribution for the "news" genre
news_freq_dist = cfd['news']
print("Most common words in 'news' genre:", news_freq_dist.most_common(10))

# Example 2: List of genres that have the word 'government'
genres_with_government = [genre for genre in brown.categories() if 'government' in cfd[genre]]
print("Genres with the word 'government':", genres_with_government)

# Example 3: Plot the frequency distribution of the word "president" across multiple genres
import matplotlib.pyplot as plt

genres = ['news', 'government', 'romance', 'humor']
counts = [cfd[genre]['president'] for genre in genres]
plt.bar(genres, counts)
plt.xlabel('Genre')
plt.ylabel('Frequency of "president"')
plt.title('Frequency of the Word "President" Across Different Genres')
plt.show()
```

```python
import nltk
from nltk.corpus import wordnet as wn

# Download WordNet
nltk.download('wordnet')

# Example 1: Finding synonyms using WordNet
synsets = wn.synsets('bank')
for synset in synsets:
    print(f"Synonym Set: {synset.name()}")
    print(f"Definition: {synset.definition()}")
    print(f"Example Sentences: {synset.examples()}")
    print()

# Example 2: Finding antonyms using WordNet
happy_synset = wn.synsets('happy')[0]
happy_lemmas = happy_synset.lemmas()
for lemma in happy_lemmas:
    antonyms = lemma.antonyms()
    if antonyms:
        print(f"Antonyms of 'happy': {[ant.name() for ant in antonyms]}")
```

```python
import nltk
from nltk.corpus import wordnet as wn

# Ensure WordNet is available
nltk.download('wordnet')

# Define two words
word1 = 'dog'
word2 = 'wolf'

# Get synsets for each word
dog_synsets = wn.synsets(word1)
wolf_synsets = wn.synsets(word2)

# Choose the first synset for each word
dog = dog_synsets[0]
wolf = wolf_synsets[0]

# Calculate Path Similarity
path_similarity = dog.path_similarity(wolf)
print(f"Path Similarity between {dog.name()} and {wolf.name()}: {path_similarity}")

# Calculate Wu-Palmer Similarity (based on the depth of the LCS)
wup_similarity = dog.wup_similarity(wolf)
print(f"Wu-Palmer Similarity between {dog.name()} and {wolf.name()}: {wup_similarity}")

# Finding the Lowest Common Subsumer (LCS)
lcs = dog.lowest_common_hypernyms(wolf)
print(f"Lowest Common Subsumer of {dog.name()} and {wolf.name()}: {lcs[0].name()}")
```

```python
from gensim.models import KeyedVectors

# Load pre-trained Word2Vec model (Google News dataset)
model_path = 'GoogleNews-vectors-negative300.bin'
model = KeyedVectors.load_word2vec_format(model_path, binary=True)

# Example: Find similar words to 'king'
similar_words = model.most_similar('king')
print("Words similar to 'king':", similar_words)

# Example: Word analogy - King is to Man as Queen is to Woman
result = model.most_similar(positive=['king', 'woman'], negative=['man'])
print("King is to Man as Queen is to Woman:", result[0])

# Example: Word similarity - Calculate cosine similarity between words
similarity = model.similarity('king', 'queen')
print("Similarity between 'king' and 'queen':", similarity)
```
